```python
df=pd.DataFrame({'A':range(1,6),
                 'B':range(6,-3,-2),
                 'C':range(10,5,-1),
                 'D clms':range(3,-2,-1),
                 'E':['python','java','python','go','python']})
df
```

```python
df.query('A>B')
```

```python
df[df.A>df.B]
```

```python
df.query('C==10')
```

```python
df.query('A>0')
```

```python
df.query('E == "python"')
```

```python
df.query('E != "python"')
```

```python
df.query('A == "D clms"')
```

```python
df[df.A == df["D clms"]]
```

```python
df.query('A == `D clms`') #tab키 위에 ` 사용
```

```python
df.query('A>3 | B>8')
```

```python
df.query('A>2 & C>B')
```

```python
df.query('E in ["python","go"]')
```

```python
num=3
df.query('B>@num')
```

```python
num=3
df.query(f'B>{num}')
```

```python
s = "E.str.contains('py')"
df.query(s)
```

```python
s = "E.str.startswith('g')"
df.query(s)
```
